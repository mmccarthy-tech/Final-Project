```python
import pandas as pd
import numpy as np

from google.colab import drive

drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Final/survey_results_public.csv')

df.head()
```

```python
#Filter Country to United States of America

df = df[df['Country'] == 'United States of America']
```

```python
#Delete rows where ConvertedCompYearly is blank

df.dropna(subset=['ConvertedCompYearly'], inplace=True)
```

```python
#Keep only Employment = "Employed"

df = df[df['Employment'] == 'Employed']
```

```python
#Remove Extreme Outliers: Keep only salaries between $30,000 and $500,000

df = df[(df['ConvertedCompYearly'] >= 30000) & (df['ConvertedCompYearly'] <= 500000)]
```

```python
#Filter by experience (YearsCodePro was not a real column, but WorkExp was)
df = df[(df['WorkExp'] >= 1) & (df['WorkExp'] <= 30)]
```

```python
#display length and width of dataset

print(df.shape[0])
print(df.shape[1])
```
